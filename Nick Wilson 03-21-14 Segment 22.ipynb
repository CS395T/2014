First, let's see if I can reproduce the $b_3 * b_5$ example from the slides using both methods

```python
# Method 1.
_, _, b3, _, b5 = b_fit
grad_f = np.array([0, 0, b5, 0, b3])
print 'standard deviation', float(np.sqrt(grad_f.dot(b_cov).dot(grad_f.T)))
```

```python
def f_product_b3_b5(b):
    _, _, b3, _, b5 = b
    return b3 * b5
```

```python
def estimate_posterior(f, b_fit, b_cov, *args, **kwargs):
    x = np.random.multivariate_normal(b_fit, b_cov, 100000)
    humps = np.apply_along_axis(f, 1, x)
    print '{0} +/- {1}'.format(np.mean(humps), np.std(humps))
    plt.hist(humps, bins=50, *args, **kwargs)
    plt.title('Posterior')
```

```python
# Method 2.
estimate_posterior(f_product_b3_b5, b_fit, b_cov)
```

Very close. I assume the slight difference is due to the truncated values I copied from the slides.
